```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 400)
y1 = np.arcsin(x + 1.2)
y2 = (2 - np.cos(x)) / 2

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = asin(x + 1.2)')
plt.plot(x, y2, label='y = (2 - cos(x)) / 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphical Solution of the System of Equations')
plt.legend()
plt.grid(True)
plt.show()
```
